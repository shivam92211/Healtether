```python
from PIL import Image
import numpy as np
```

```python
def boolean_array_to_image(boolean_array, output_path="output_image.png"):

   # Convert boolean array to integer array (0s and 1s)
   binary_array = np.array(boolean_array, dtype=np.uint8) * 255

  # Reshape the array into a 2D array (assuming it represents a grayscale image)
   height, width = binary_array.shape
   image_array = binary_array.reshape((height, width))


   # Create an image from the array
   image = Image.fromarray(image_array, mode='L')  # 'L' mode for grayscale


   # Save the image to a file
   image.save(output_path)
   image.show(output_path)
```

```python
boolean_array = np.array([[0, 0, 0],
                         [1, 0, 1],
                         [0, 0, 0]], dtype=bool)


boolean_array_to_image(boolean_array, output_path="output_image.png")
```

```python
# Assuming ima_gray is your grayscale image as a NumPy array
ima_gray = cv2.cvtColor(new_ima, cv2.COLOR_BGR2GRAY)

threshold = 127

# Thresholding the grayscale image to create a binary image
bin_img = np.zeros_like(ima_gray)
bin_img[ima_gray > threshold] = 255  # Set pixels above threshold to 255 (white)

# Convert the binary array to an image
bin_image = Image.fromarray(bin_img.astype(np.uint8))

# Display the binary image using matplotlib
plt.imshow(bin_image, cmap='gray')
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
# Conditionally set elements to 1 if equal to 255
gray_arr[gray_arr == 255] = 1

print(gray_arr)
```

```python
import csv

csv_file_name = 'binary_data.csv'

with open(csv_file_name, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(gray_arr)
```

```python
# Define the CSV file name
csv_file_name = 'binary_data.csv'

# Load data from the CSV file
loaded_data = []
with open(csv_file_name, 'r', newline='') as file:
    reader = csv.reader(file)
    for row in reader:
        # Convert each row to integers and append to loaded_data
        loaded_data.append([int(val) for val in row])

# Convert loaded data to a nested NumPy array
nested_array = np.array(loaded_data)

print(nested_array)
```

```python
boolean_array_to_image(nested_array, output_path="output_image3.png")
```

```python
import GPUtil

def get_gpu_info():
    try:
        # Get the list of available GPUs
        gpus = GPUtil.getGPUs()

        if not gpus:
            print("No GPU found.")
        else:
            for i, gpu in enumerate(gpus):
                print(f"GPU {i + 1}:")
                print(f"  Name: {gpu.name}")
                print(f"  ID: {gpu.id}")
                print(f"  Driver: {gpu.driver}")
                print(f"  GPU Memory Total: {gpu.memoryTotal} MB")
                print(f"  GPU Memory Free: {gpu.memoryFree} MB")
                print(f"  GPU Memory Used: {gpu.memoryUsed} MB")
                print(f"  GPU Load: {gpu.load * 100}%")
                print()

    except Exception as e:
        print(f"Error: {e}")

# Call the function to get GPU information
get_gpu_info()
```
